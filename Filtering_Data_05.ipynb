```python
houses[(houses['waterfront']== 1) & (houses['price'] < 500000)].sort_values(by='price',ascending=False)
```

```python
houses[(houses['view'] == 4) & (houses['grade'] >= 11) & (houses['sqft_living'] <= 3000)]
```

```python
high_quality = houses['grade'] >= 11
good_view = houses['view'] == 4
smaller = houses['sqft_living'] <= 3000
houses[high_quality & good_view & smaller]
```

```python
# OR operator (|)
```

```python
houses[houses['yr_built'] >= 2014]
houses[houses['yr_renovated'] >= 2014]
recent = houses[(houses['yr_built'] >= 2014) |(houses['yr_renovated'] >= 2014) ]
```

```python
recent.sort_values(by='yr_renovated',ascending=False)
```

```python
netflix[(netflix['director'] == 'David Fincher') | (netflix['director'] == 'Martin Scorsese')]
```

```python
netflix[(netflix['director'] == 'David Fincher') | (netflix['director'] == 'Martin Scorsese')]['director'].value_counts().plot(kind='bar')
```

```python
netflix[netflix['director'].isin(['David Fincher','Martin Scorsese'])]
```

```python
david = netflix['director'] == 'David Fincher'
martin = netflix['director'] == 'Martin Scorsese'
recent =  netflix['release_year'] > 2015
netflix[(david | martin) & recent]
```
