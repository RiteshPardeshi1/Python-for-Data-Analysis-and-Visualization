```python
ufos['months'] = ufos['date_time'].dt.month
```

```python
sightings = ufos['months'].value_counts().sort_index()
sightings.plot(
    kind='bar',
    title='Ufo with respect to months',
    xlabel="Months", ylabel="Nums sightings", 
    fontsize=10)
```

```python
sightings = ufos['months'].value_counts().sort_index()
sightings.plot(
    kind='bar',
    title='Ufo with respect to months',
    fontsize=10)
plt.xlabel("Month",fontsize=30,color='olive')
```

```python
months_dict = {1: "Jan", 2: "Feb", 3: "Mar", 4: "Apr", 5: "May", 6: "Jun", 7:"Jul", 8: "Aug", 9:"Sep", 10: "Oct", 11: "Nov", 12: "Dec"}
sightings.rename(months_dict)
```

```python
sightings.rename(months_dict).plot(kind='bar')
```

```python
sightings.rename(months_dict).plot(
    kind='bar',
    title='Ufo with respect to months',
    fontsize=10)
plt.xlabel("Month",fontsize=30,color='olive')
```

```python
ufos[ufos['state']=="CA"].year.value_counts().sort_index().plot()
ufos[ufos['state']=="TX"].year.value_counts().sort_index().plot()
```

```python
ufos[ufos['state']=="CA"].year.value_counts().sort_index().plot(color='olive')
plt.figure()
ufos[ufos['state']=="TX"].year.value_counts().sort_index().plot()
```

```python
ufos[ufos['state']=="CA"].year.value_counts().sort_index().plot(label='CA')
ufos[ufos['state']=="TX"].year.value_counts().sort_index().plot(label='TX')
plt.legend()
plt.title('State wise UFOs sightings')
plt.xlabel("Years")
```

```python
plt.figure(figsize=(10,8))
ufos[ufos['shape'] == 'light']['year'].value_counts().sort_index().plot(label='light')
ufos[ufos['shape'] == 'fireball']['year'].value_counts().sort_index().plot(label='fireball')
ufos[ufos['shape'] == 'circle']['year'].value_counts().sort_index().plot(label='circle')
ufos[ufos['shape'] == 'triangle']['year'].value_counts().sort_index().plot(label='triangle')
ufos[ufos['shape'] == 'formation']['year'].value_counts().sort_index().plot(label='formation')
plt.xlim(2000,2018)
plt.legend()
plt.title("UFOs Sightings By Shape")
```

```python
fig, axs = plt.subplots(1,2)
ufos.year.value_counts().sort_index().plot(ax=axs[0])
axs[0].set_title("Sighting by year")
ufos.months.value_counts().sort_index().plot(kind='bar',ax=axs[1])
axs[1].set_title("Sighting by month")
```

```python
months = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 
          7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
```

```python
fig, axs = plt.subplots(2,3, figsize=(14,8))
plt.suptitle("UFOs Sighting by month",fontsize= 14)
ufos[ufos['year'] == 2014]['months'].value_counts().sort_index().rename(months).plot(kind='bar',ax=axs[0][0],title="2014",color='olive')
ufos[ufos['year'] == 2015]['months'].value_counts().sort_index().rename(months).plot(kind='bar',ax=axs[0][1],title="2015",color='red')
ufos[ufos['year'] == 2016]['months'].value_counts().sort_index().rename(months).plot(kind='bar',ax=axs[0][2],title="2016",color='blue')
ufos[ufos['year'] == 2017]['months'].value_counts().sort_index().rename(months).plot(kind='bar',ax=axs[1][0],title="2017",color='pink')
ufos[ufos['year'] == 2018]['months'].value_counts().sort_index().rename(months).plot(kind='bar',ax=axs[1][1],title="2018",color='green')
ufos[ufos['year'] == 2019]['months'].value_counts().sort_index().rename(months).plot(kind='bar',ax=axs[1][2],title="2019",color='black')
plt.tight_layout()
```
