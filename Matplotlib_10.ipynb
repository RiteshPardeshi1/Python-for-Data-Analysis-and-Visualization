```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10,25,5,30,21]
germinated = [74, 88, 56,69,59]
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,germinated,width=0.5,color='peru')
plt.title("Germination rate")
plt.ylabel("Number of Seedings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,died,width=0.5,color='orchid')
plt.title("Failure rate")
plt.ylabel("Number of Seedings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,died,width=0.5,color='orchid')
plt.title("Failure rate")
plt.ylabel("Number of Seedings")
plt.figure(figsize=(6,6))
plt.bar(plants,germinated,width=0.6)
plt.title("Germination rate")
plt.ylabel("Number of Seedings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db")
plt.bar(plants, germinated, width=0.4, color="#2ecc71", align="edge")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db",bottom=germinated,label="Failed Seedlings")
plt.bar(plants, germinated,color="#2ecc71",label="Germinated Seedling")
plt.legend(frameon=True,shadow=True,facecolor="white")
plt.title("Greenhouse Seedlings")
```

```python
plt.figure(figsize=(6,6))
plt.barh(plants, died, color="#3498db")
```

```python
nums = np.arange(5)
plt.figure(figsize=(10,5))
plt.suptitle("Our First Subplot",fontsize=30)
plt.subplot(1,3,1)
plt.plot(nums,nums)
plt.title("x")

plt.subplot(1,3,2)
plt.bar(nums,nums*4,color='green')
plt.title("X*4")

plt.subplot(1,3,3)
plt.scatter(nums,nums**3,color="olive")
plt.title("X cube")

plt.suptitle("Our First Subplot")
plt.tight_layout()
plt.show()
```

```python
first_class = titanic[titanic["pclass"] == 1]["age"]
second_class = titanic[titanic["pclass"] == 2]["age"]
third_class = titanic[titanic["pclass"] == 3]["age"]

plt.figure(figsize=(10,4))
ax = plt.subplot(1,3,1)
plt.hist(first_class, label="1st Class", color="#f1c40f", bins=20)
plt.title("1st Class")

plt.subplot(1,3,2, sharey=ax)
plt.hist(second_class, label="2nd Class", color="#3498db", bins=20)
plt.title("2nd Class")


plt.subplot(1,3,3, sharey=ax)
plt.hist(third_class, label="3rd Class", color="#e74c3c", bins=20)
plt.title("3rd Class")
```
